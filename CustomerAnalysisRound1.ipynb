```python
file1.columns = file1.columns.str.replace(' ','_')
file2.columns = file2.columns.str.replace(' ','_')
file3.columns = file3.columns.str.replace(' ','_')

cols1 = []
for column in file1.columns:
    cols1.append(column.lower())
    
file1.columns = cols1

cols2 = []
for column in file2.columns:
    cols2.append(column.lower())
file2.columns = cols2

cols3 = []
for column in file3.columns:
    cols3.append(column.lower())
file3.columns = cols3


```

```python
file3 = file1.rename(columns={'state':'st'})
file1 = file1[['customer', 'customer_lifetime_value', 'education', 'gender', 'income','monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'st', 'total_claim_amount', 'vehicle_class']]
file2 = file2[['customer', 'customer_lifetime_value', 'education', 'gender', 'income','monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'st', 'total_claim_amount', 'vehicle_class']]
file3 = file3[['customer', 'customer_lifetime_value', 'education', 'gender', 'income','monthly_premium_auto', 'number_of_open_complaints', 'policy_type', 'st', 'total_claim_amount', 'vehicle_class']]

data = pd.concat([file1, file2, file3], axis=0)
```

```python
data[:100]
```

```python
# numerical columns
data.select_dtypes(include=["integer", "float"])
```

```python
# categorical columns
data.select_dtypes(include=["object"])
```

```python
data.drop('number_of_open_complaints', axis=1, inplace=True)
data.drop('education', axis=1, inplace=True)
```

```python
data["customer_lifetime_value"].unique()

data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace('%','')
```

```python
data['customer_lifetime_value'] =  pd.to_numeric(data['customer_lifetime_value'], errors='coerce')
data["customer_lifetime_value"].unique()
```

```python
data = data.drop_duplicates()
```
